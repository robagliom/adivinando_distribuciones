```python
#Librerías necesarias:
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import math
```

```python
#Importamos datos
datos_01 = pd.read_excel('Datos01.xlsx', header = None)
lista_01 = {col:datos_01[col].dropna().tolist() for col in datos_01.columns}[0]
datos_02 = pd.read_excel('Datos02.xlsx')
lista_02 = {col:datos_02[col].dropna().tolist() for col in datos_02.columns}['Val. Distribuidos…(?)']
datos_03 = pd.read_excel('Datos03.xlsx', header = None)
lista_03 = {col:datos_03[col].dropna().tolist() for col in datos_03.columns}[0]
```

```python
#Histograma datos_01
print(lista_01[:10]) #Vemos que son datos discretos
plt.hist(lista_01,alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.title('HISTOGRAMA: datos_01')  # Colocamos el titulo
plt.xlabel('intervalos de clases') # Nombre x
plt.ylabel('frecuencia absoluta') # Nombre y
plt.show()
plt.clf()

#El histograma tiene forma de binomial; descartamos la distribución normal ya que los datos son discretos
#Vamos a comparar con los valores teóricos
#Esperanza de la muestra
esperanza = statistics.mean(lista_01)
#Varianza de la muestra
varianza = statistics.pvariance(lista_01)
print('Esperanza en la muestra: ', esperanza)
print('Varianza en la muestra: ',varianza)

#Calculamos valores teóricos n y p
p = (esperanza-varianza)/esperanza
n = (esperanza**2)/(esperanza - varianza)
print('p: ',p,' - n: ', n)
#Calculamos esperanza y varianza con n y p
esperanza_teorica = n*p
varianza_teorica = n*p*(1-p)
print('Esperanza teorica: ', esperanza_teorica)
print('Varianza teorica: ',varianza_teorica)
#Los valores de la muestra son muy parecidos a los téoricos, por lo tanto se corresponde con esa distribución
print('La distribución es Binomial!!')
```

```python
#Histograma datos_02
print(lista_02[:10]) #Vemos que los datos continuos
plt.hist(lista_02,alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.title('HISTOGRAMA: datos_02')  # Colocamos el titulo
plt.xlabel('intervalos de clases') # Nombre x
plt.ylabel('frecuencia absoluta') # Nombre y
plt.show()
plt.clf()

#El histograma tiene forma de exponencial, como los datos son continuos es posible esta distribución
#Vamos a comparar con los valores teóricos
#Esperanza de la muestra
esperanza = statistics.mean(lista_02)
#Varianza de la muestra
varianza = statistics.pvariance(lista_02)
print('Esperanza en la muestra: ', esperanza)
print('Varianza en la muestra: ',varianza)

#Calculamos alfa
alfa = 1/esperanza
print('alfa: ',alfa)
#Calculamos valores teóricos
esperanza_teorica = 1/alfa
varianza_teorica = 1/alfa**2
print('Esperanza teorica: ', esperanza_teorica)
print('Varianza teorica: ',varianza_teorica)
#Los valores de la muestra son muy parecidos a los téoricos, por lo tanto se corresponde con esa distribución
print('La distribución es exponencial!!!')
```

```python
#Histograma datos_03
print(lista_03[:10]) #Vemos que los datos son discretos
plt.hist(lista_03,alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.title('HISTOGRAMA: datos_03')  # Colocamos el titulo
plt.xlabel('intervalos de clases') # Nombre x
plt.ylabel('frecuencia absoluta') # Nombre y
plt.show()
plt.clf()

#El histograma tiene forma de poisson, como los datos son discretos es probable esta distribución
#Esperanza de la muestra
esperanza = statistics.mean(lista_03)
#Varianza de la muestra
varianza = statistics.pvariance(lista_03)
print('Esperanza en la muestra: ', esperanza)
print('Varianza en la muestra: ',varianza)

#Como la distribución es discreta y la gráfica se asemeja a una distribución de Poisson, concluimos
#que esta muestra se distribuye Poisson
print("La distribución es de Poisson!!!")
```
